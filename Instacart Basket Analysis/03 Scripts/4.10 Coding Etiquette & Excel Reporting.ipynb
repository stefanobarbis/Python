### Question 3: Regions Spending Trends

```python
region = []

for x in opcf["STATE"]:
    if x =="Maine" or x == "New Hampshire" or x == "Vermont" or x == "Massachusetts" or x == "Rhode Island" or x == "Connecticut" or x == "New York" or x == "Pennsylvania" or x == "New Jersey":
        region.append("Northeast")
    elif x == "Wisconsin" or x == "Michigan" or x == "Illinois" or x == "Indiana" or x == "Ohio" or x == "North Dakota" or x == "South Dakota" or x == "Nebraska" or x == "Kansas" or x == "Minnesota" or x == "Iowa" or x == "Missouri":
        region.append("Midwest")
    elif x == "Delaware" or x == "Maryland" or x == "District of Columbia" or x== "Virginia" or x == "West Virginia" or x == "North Carolina" or x == "South Carolina" or x == "Georgia" or x == "Florida" or x == "Kentucky" or x == "Tennessee" or x == "Mississippi" or x == "Alabama" or x == "Oklahoma" or x == "Texas" or x == "Arkansas" or x == "Louisiana":
        region.append("South")
    else:
        region.append("West")
    
  
```

```python
opcf['Region'] = region
```

```python
opcf.head()
```

```python
crosstab = pd.crosstab(opcf['Region'], opcf['loyalty_flag'], dropna = False)
```

### Question 4: Exclusion flag

```python
frequency_flag = []

for x in opcf["add_to_cart_order"]:
    if x < 5:
        frequency_flag.append("Exclude")
    else:
        frequency_flag.append("Include")
```

```python
opcf['Frequency_flag'] = frequency_flag
```

```python
opcf_frequent =  opcf[opcf['Frequency_flag']== 'Include']
```

### Question 7: Aggregation

```python
opcf_frequent.groupby('status').agg({'days_since_prior_order': ['mean', 'min', 'max']})
```

```python
opcf_frequent.groupby('status').agg({'prices': ['mean', 'min', 'max']})
```

Interestingly, we can see that the wealthy income group comes in a more frequently basis. Furthermore, we can see that the wealthy group tends to buy more expensive products as well.

```python
opcf_frequent.groupby('Age_group').agg({'days_since_prior_order': ['mean', 'min', 'max']})
```

```python
opcf_frequent.groupby('Age_group').agg({'prices': ['mean', 'min', 'max']})
```

On this aggregation, we note that the mature group tend to buy more frequent as well as higher prices. 

Overall, there seems to be a correlation with age and income that we must take into account since they are our most profitable buyer.
