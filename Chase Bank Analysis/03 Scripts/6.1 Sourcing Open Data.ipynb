### Data Profile

```python
df['Acquired Date'].value_counts(dropna = False)
```

Because missing values (NaN) do not represent 90-95% of the data, we will not delete this row. However, we will drop the rows with this missing values.

```python
df_clean = df[df['Acquired Date'].isnull() == False]
```

```python
df_clean['Acquired Date'].value_counts(dropna = False)
```

```python
for col in df_clean.columns.tolist():
  weird = (df_clean[[col]].applymap(type) != df_clean[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_clean[weird]) > 0:
    print (col)
```

According to the output above, we do not have any mixed type columns

```python
df_clean.isnull().sum()
```

According to the above output we have some missing data, we will delete those column with missing deposits.

```python
df_clean = df_clean[df_clean['2010 Deposits'].isnull() == False]
```

```python
df_clean = df_clean[df_clean['2011 Deposits'].isnull() == False]
```

```python
df_clean.isnull().sum()
```
